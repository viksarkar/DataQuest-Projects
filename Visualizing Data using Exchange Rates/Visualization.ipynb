## Guided Project: Storytelling Data Visualization on Exchange Rates

This project will use a dataset of the daily Euro-US Dollar exchange rate between 1999 and 2021. If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020) — the data source is the European Central Bank. Note that the dataset gets regular updates — This particular dataset was downloaded on January 2021.

We will import the data and look at the first and last five rows to get an idea of what the data is. 

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
print(exchange_rates.describe())
print(exchange_rates.info())
```

Looks like there are 5698 rows of data with exchange rates against 39 different currencies between Jan 4 1999 and Jan 8 2021. There are null values and most of the values are floats and objects, except for one date/time column.

For this project, we will be focusing on the exchange rate against the US Dollar specifically. To help, we will rename relevant columns and sort the data by time. 

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

Looks like there are 62 non numeric values in this series. These dashes represent days when there is no available data for the exchange rate. We will select all rows where there are no dashes and also convert all the values to float. 

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar']!='-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

So looks like we are down to 5637 data points and all the exchange rates are floats like we wanted. 

Now let us generate a line plot to visualize the evolution of the euro-dollar exchange rate. 

```python
import matplotlib.pyplot as plt
## The following line enables Jupyter to display graphs
%matplotlib inline 

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

Looking at this graph, it is clear that there are two trends overlaid on each other. There is a daily trend of ups and downs and a longer term trend. Sometimes it is advantageous to just see the long term trend. We can do this by using the pandas.Series.rolling().mean() method. Let us do that to see the mean with a rolling window of 30 days. 

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar.rolling(30).mean()
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

So now we have to come up with an interesting way to use this data and create a visualization. Since we have data over the course of three US presidencies, it will be interesting to see how things changed with each administration (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)).

```python
# First let's get the data subset covering all three presidents
# and then let us break it down further for each president. 
euro_dollar_BOT = euro_to_dollar[(euro_to_dollar['Time'].dt.year>=2001)]
euro_dollar_GB = euro_dollar_BOT[(euro_to_dollar['Time'].dt.year<2009)]
euro_dollar_BO = euro_dollar_BOT[(euro_to_dollar['Time'].dt.year>=2009) &
                                 (euro_to_dollar['Time'].dt.year<2017)]
euro_dollar_DT = euro_dollar_BOT[(euro_to_dollar['Time'].dt.year>=2017)]

### We will be using the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

```

```python
plt.figure(figsize=(8,4))
fig, ax = plt.subplots()
plt.plot(euro_dollar_GB['Time'], euro_dollar_GB['rolling_mean'], color='red', linewidth = 0.75)
plt.plot(euro_dollar_BO['Time'], euro_dollar_BO['rolling_mean'], color='blue', linewidth = 0.75)
plt.plot(euro_dollar_DT['Time'], euro_dollar_DT['rolling_mean'], color='red', linewidth = 0.75)
ax.axvline(x=pd.to_datetime('2009-01-01'), linestyle = '--', linewidth = 0.5, color='black')
ax.axvline(x=pd.to_datetime('2017-01-01'), linestyle = '--', linewidth = 0.5, color='black')
ax.set_xticks([pd.to_datetime('2001-01-01'), pd.to_datetime('2005-01-01'), pd.to_datetime('2009-01-01'),
               pd.to_datetime('2013-01-01'), pd.to_datetime('2017-01-01'), pd.to_datetime('2021-01-01')])
ax.set_xticklabels(['2001','2005','2009','2013','2017','2021'], alpha=0.5)
ax.set_ylim(0.8, 1.6)
ax.set_yticks([0.8, 1.0, 1.2, 1.4, 1.6])
ax.set_yticklabels([0.8, 1.0, 1.2, 1.4, 1.6],alpha=0.5)
ax.text(pd.to_datetime('1999-01-01'), 1.7, '''EUR-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''', fontsize=10)
ax.text(pd.to_datetime('2003-01-01'), 1.6, 'George W Bush', fontsize=10)
plt.show()
```
